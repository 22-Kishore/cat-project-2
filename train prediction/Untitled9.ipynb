```python
import pandas as pd
import numpy as np

# Load the dataset into a pandas DataFrame
df = pd.read_csv("E:\\6thsem\\data mining\\All_Indian_Trains.csv")

# Print the first five rows of the DataFrame
print(df.head())
```

```python
# Check the data types of each column
print(df.dtypes)
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Print summary statistics
print(df.describe())
```

```python
import matplotlib.pyplot as plt
# Create a bar chart of the Starts column
starts_count = df['Starts'].value_counts()
plt.bar(starts_count.index, starts_count.values)
plt.title('Number of Trains Starting at Each Station')
plt.xlabel('Station')
plt.ylabel('Number of Trains')
plt.show()
```

```python
train_name_count = df['Train name'].value_counts()

# Print the count of unique train names
print(train_name_count)
```

```python
plt.figure(figsize=(100,10))

# Plot the result as a bar chart
plt.bar(train_name_count.index, train_name_count.values)
plt.title("Train Name Distribution")
plt.xlabel("Train Name")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```

```python
grouped_data = df.groupby(['Starts', 'Ends']).size().reset_index(name='Count')

# Print the grouped data
print(grouped_data)
```
